```python
def checkYearMonth(row):
    value = row['date']
    return str(value)[0:6]
```

```python
df['yearmonth'] = df.apply(checkYearMonth, axis=1)
```

Number of news related to covid in timeline

```python
df4 = df['yearmonth'].value_counts()
df4
```
